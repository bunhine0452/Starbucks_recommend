```python
import pandas as pd

# 두 개의 파일 경로를 지정합니다.
file_0 = "starbucks_naver_blog_batch_18.csv"
file_1 = "starbucks_naver_blog_batch_19.csv"

# 두 파일을 읽어옵니다.
df_0 = pd.read_csv(file_0, error_bad_lines=False, engine='python')
df_1 = pd.read_csv(file_1, error_bad_lines=False, engine='python')

# None 값을 빈 문자열로 대체합니다.
df_0['Content'] = df_0['Content'].fillna('')
df_1['Content'] = df_1['Content'].fillna('')

# 두 데이터프레임을 합칩니다.
combined_df = pd.concat([df_0, df_1], ignore_index=True)

# Store_Name이 같은 Content를 하나의 열로 합칩니다.
combined_df['Content'] = combined_df.groupby('Store_Name')['Content'].transform(lambda x: ' '.join(x))
combined_df = combined_df.drop_duplicates(subset=['Store_Name'])

# 합친 데이터를 csv 파일로 저장합니다.
output_file = "./output/combined_starbucks_blog_batch_18_19.csv"
combined_df.to_csv(output_file, index=False)

# 결과 파일 경로를 반환합니다.
print("결과 파일 저장 완료:", output_file)
```
