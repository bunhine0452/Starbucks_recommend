```python
import os
import time
import re
import requests
import pandas as pd
import urllib.parse
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#네이버 api 키를 발급 후 입력해주세요 

def get_blog_urls(store_name):
    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret,
    }
    encText = urllib.parse.quote(store_name + " 스타벅스")
    url = f"https://openapi.naver.com/v1/search/blog.json?query={encText}&display=10"
    print(f"Requesting URL: {url}")
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()
        urls = [item['link'] for item in data['items'] if 'blog.naver.com' in item['link']]
        print(f"Found URLs for '{store_name}': {urls}")
        return urls
    else:
        print(f"Error Code: {response.status_code}")
        return []

def main():
    # 스타벅스 매장 데이터를 불러옵니다.
    starbucks_data = pd.read_csv('./data/starbucks_data.csv')
    blog_data = []

    for store_name in starbucks_data['Store_Name']:
        print(f"Processing store: {store_name}")
        urls = get_blog_urls(store_name)
        for url in urls:
            blog_data.append({'Store_Name': store_name, 'URL': url})
        print(f"Collected URLs for '{store_name}': {urls}")

    # 수집한 블로그 URL 데이터를 CSV 파일로 저장합니다.
    blog_df = pd.DataFrame(blog_data)
    blog_df.to_csv('naver_blog_starbucks.csv', index=False, encoding='utf-8-sig')
    print("Saved collected URLs to 'naver_blog_starbucks.csv'")

if __name__ == '__main__':
    main()
```
